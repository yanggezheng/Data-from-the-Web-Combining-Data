```python
from bs4 import BeautifulSoup
with open('./COURSE SCHEDULE.html') as file:
    lines = file.readlines()[16:]
html = ''.join(lines)
dom = BeautifulSoup(html)
dom
```

```python
schedule = dom.select_one('.schedule-listing')
courses = schedule.select('li')
courses
```

```python
import pandas as pd
df = pd.DataFrame(columns=['Number-Section','Name','Instructor','Time'])
for course in courses:
    section = course.select('span')[0].text.strip()
    name = course.select('span')[1].text.strip()
    instructor = course.select('span')[2].text.strip()
    time = course.select('span')[3].text.strip()
    course_dict = {'Number-Section': section, 'Name': name, 'Instructor': instructor, 'Time': time}
    df = df.append(course_dict, ignore_index=True)
df
```

```python
df['Number-Section'] = df['Number-Section'].apply(lambda x:x.split('\n')[0])
df
```

For courses with 2 course numbers, I ignored the second one

```python
df['Name'] = df['Name'].apply(lambda x:'Special Topics: '+x.split('Topics:')[1].strip() if 'Special Topics:'in x else x)
df
```

I removd 'Special Topics:' and trim the string, then add 'Special Topics:' back

```python
df
```

```python
import re
df['Number'] = df['Number-Section'].apply(lambda x : re.search('(\w{4}-\w{2}.\d{4})',x)[1])
df['Section'] = df['Number-Section'].apply(lambda x : re.search('-​(\d{3})',x)[1])
df
```

```python
from bs4 import BeautifulSoup
with open('./COURSE CATALOG.html') as file:
    lines = file.readlines()
html1 = ''.join(lines)
dom1 = BeautifulSoup(html1)
dom1
```

```python
schedule = dom1.select_one('.courses-listing')
courses = schedule.select('li')
courses
```

```python
df1 = pd.DataFrame(columns=['Number','Prereqs','Points'])
for course in courses:
    number = course.select('p')[0].text.strip().split(' ')[0]
    points = course.select('p')[1].text.strip().split('Points')[0]
    pre = course.select('p')[2].text.split('Prerequisites: ')[1]
    course_dict = {'Number': number, 'Points': points, 'Prereqs': pre}
    df1 = df1.append(course_dict, ignore_index=True)
df1
```
