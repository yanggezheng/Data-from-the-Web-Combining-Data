```python
import pandas as pd
df = pd.DataFrame(columns=['Director', 'rt_score'])
for i in range(1,4):
    url2 = f'http://linserv1.cims.nyu.edu:10000/films?_page={i}'
    res = requests.get(url2)
    d = res.json()
    for movie in d:
        director = movie.get('director')
        score = movie.get('rt_score')
        movie_dict = {'Director': director, 'rt_score': score}
        df = df.append(movie_dict, ignore_index=True)
df['rt_score'] = df['rt_score'].astype('Float64')
df
```

```python
scores = df.groupby('Director')['rt_score'].mean()
count = df.groupby('Director')['rt_score'].count()
result = pd.concat([scores, count], axis=1)
result.columns = ['avg_rt_score', 'count']
result.index.name = 'Director'
result.sort_values(by = 'avg_rt_score',ascending=False)
```
